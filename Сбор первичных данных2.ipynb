```python
def Map(part_hrefs):
    # В эту табличку будем собирать данные
    dfr = pd.DataFrame( )
    k = 0 # Это номера наблюдений
    n = len(part_hrefs)
    for item in part_hrefs:
        k = k + 1
        # грузим новое наблюдение/
        dfr1 = pd.DataFrame.from_dict(One_Flat(item),orient='index')
        # присваиваем этому наблюдению номер
        dfr1.columns =[k]
        # закидываем его в итоговую таблицу
        dfr = dfr.join(dfr1, how='outer')
        # Выдавать информацию о том, что сделана каждая десятая итерация! 
        if k%100 == 0:
            print('сделано ', k, ' итераций из', n)
    return(dfr)
```

```python
def One_Flat(href):
    data = { }  # Задали пустой словарь, в который мы будем сохранять данные
    data['link'] = href
    # Подгружаем страничку с информацией по квартире
    response = requests.get(href)
    html = response.content
    soup = BeautifulSoup(html,'html.parser')
    
    # Вытаскиваем цену на квартиру
    try:
        l = soup.findAll('span', itemprop="price")
        price = l[0].attrs["content"].replace(' ','')
        data['Цена'] = price
    except Exception:
        data['Цена'] = 'NA'
    
    # Вытаскиваем расстояние до метро то ли пешком то ли на машине
    try:
        Do_metro = soup.findAll('span', attrs = {'class':"a10a3f92e9--underground_time--1fKft"})[0].text.strip()
        data['До Метро'] = Do_metro
    except Exception:
        data['До Метро'] = "NA"
    
    
    # Вытаскиваем количество комнат в квартире
    try:
        komnata=soup.findAll('h1', attrs = {'class':"a10a3f92e9--title--2Widg"})[0].text
        data['Комнаты'] = komnata
    except Exception:
        data['Комнаты'] = 'NA'
    # Вытаскиваем район квартиры
    try:
        raion = soup.findAll('a', attrs = {'class':"a10a3f92e9--link--1t8n1 a10a3f92e9--address-item--1clHr"})[2].text
        data['Район'] = raion
    except Exception:
        data['Район'] = 'NA'

    table3=soup.findAll('div', attrs = {'class':"a10a3f92e9--info--3XiXi"})
    if len(table3)>0:
        for row in table3:
            cols3=[row.findAll('div')[1].text,row.findAll('div')[0].text ]
            data[cols3[0]] = cols3[1]
    
    try:
        descr = soup.findAll('p', itemprop="description")[0].text
        data['Описание']=descr
    except Exception:
        data['Описание'] = 'NA'
        
    type_seller=[soup.findAll('div', attrs = {'class':"a10a3f92e9--prop--1F4GC"}),
    soup.findAll('h2', attrs = {'class':"a10a3f92e9--title--BarZi"})] 
    if type_seller[0]==[]:
        if type_seller[1]==[]:
            seller='NA'
        else:
            seller=type_seller[1][0].text
    else:
        seller=type_seller[0][0].text
    data['Тип продавца'] =seller
    
     #Общие характеристики
    table =soup.findAll('section',attrs = {'class':"a10a3f92e9--container--1MHfF"})
    if len(table)>0:
        for row in table:
            cols=[row.findAll('span')[0].text.strip(),row.findAll('span')[1].text.strip() ]
            data[cols[0]] = cols[1]
            
    table1 = soup.findAll('div',attrs = {'class':"a10a3f92e9--item--2Ig2y"})
    if len(table1)>0:
        for row in table1:
            cols2=[row.findAll('div')[0].text,row.findAll('div')[1].text]
            data[cols2[0]] = cols2[1]
    return(data)
```
